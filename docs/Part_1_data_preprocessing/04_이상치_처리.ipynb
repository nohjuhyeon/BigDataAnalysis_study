# 데이터 준비

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data = 10*np.random.randn(200)+50
df= pd.DataFrame({'값':data})
df.loc[201]=2
df.loc[202]=100
df.loc[203]=10
df.loc[204]=110
```

```python
plt.hist(df['값'],bins=20,rwidth=0.8)
plt.show()
```

# IQR 값 구하기

```python
plt.boxplot(df['값'])
plt.show()
```

```python
Q1 = df['값'].quantile(.25)
Q2 = df['값'].quantile(.5)
Q3 = df['값'].quantile(.75)
IQR = Q3- Q1
print("Q1 = ",Q1)
print("Q2 = ",Q2)
print("Q3 = ",Q3)
print("IQR = ",IQR)
```

```python
Q1 = df['값'].describe()['25%']
Q2 = df['값'].describe()['50%']
Q3 = df['값'].describe()['75%']
IQR = Q3- Q1
print("Q1 = ",Q1)
print("Q2 = ",Q2)
print("Q3 = ",Q3)
print("IQR = ",IQR)
```

# 이상치 확인하기

```python
condition = df['값'] > (Q3 + IQR * 1.5)
upperOutlier = df[condition]
condition = df['값'] < (Q1 - IQR * 1.5)
lowerOutlier = df[condition]
print(upperOutlier)
print(lowerOutlier)
```
