# 데이터 준비

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv("https://raw.githubusercontent.com/YoungjinBD/dataset/main/lynx.csv")
print(df.head(10))
print(df.describe())
```

# 단순 이동 평균

```python
# 10년 단순이동평균
df['sma'] = df['value'].rolling(10).mean()
plt.plot(df['value'])
plt.plot(df['sma'])
plt.show()
```

# 지수가중이동평균

```python
# 10년 지수가중이동평균
df['ewm'] = df['value'].ewm(10).mean()
plt.plot(df['value'])
plt.plot(df['ewm'])
plt.show()
```

# 단순이동평균과 지수가중이동평균의 차이
## - 최근 데이터 반영: EMA는 최근 데이터에 더 큰 가중치를 두어 변화에 더 민감하게 반응함 반면, SMA는 모든 데이터 포인트에 동일한 가중치를 부여하여 변화에 덜 민감함
## - 계산 복잡성: SMA는 계산이 단순하지만, EMA는 매번 이전 값을 참조해야 하므로 조금 더 복잡함
## - 반응 속도: EMA는 급격한 변동을 빠르게 반영하는 반면, SMA는 변동을 반영하는 데 시간이 더 걸림
