## 이직여부 판단 데이터

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler, LabelEncoder 
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, f1_score, precision_score, roc_auc_score, recall_score
from sklearn.ensemble import RandomForestClassifier
#데이터 로드
x_train = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/HRdata/X_train.csv")
y_train = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/HRdata/y_train.csv")
x_test= pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/HRdata/X_test.csv")
y_test= pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/HRdata/y_test.csv")

print(x_train.describe(include='all'))

# 데이터 전처리

## 라벨 인코딩
x_train['gender'] = LabelEncoder().fit_transform(x_train['gender'])
x_train['relevent_experience'] = LabelEncoder().fit_transform(x_train['relevent_experience'])
x_train['enrolled_university'] = LabelEncoder().fit_transform(x_train['enrolled_university'])
x_train['education_level'] = LabelEncoder().fit_transform(x_train['education_level'])
x_train['major_discipline'] = LabelEncoder().fit_transform(x_train['major_discipline'])
x_train['company_size'] = LabelEncoder().fit_transform(x_train['company_size'])
x_train['company_type'] = LabelEncoder().fit_transform(x_train['company_type'])
x_train['last_new_job'] = LabelEncoder().fit_transform(x_train['last_new_job'])

x_test['gender'] = LabelEncoder().fit_transform(x_test['gender'])
x_test['relevent_experience'] = LabelEncoder().fit_transform(x_test['relevent_experience'])
x_test['enrolled_university'] = LabelEncoder().fit_transform(x_test['enrolled_university'])
x_test['education_level'] = LabelEncoder().fit_transform(x_test['education_level'])
x_test['major_discipline'] = LabelEncoder().fit_transform(x_test['major_discipline'])
x_test['company_size'] = LabelEncoder().fit_transform(x_test['company_size'])
x_test['company_type'] = LabelEncoder().fit_transform(x_test['company_type'])
x_test['last_new_job'] = LabelEncoder().fit_transform(x_test['last_new_job'])

Max = x_train['training_hours'].describe()['75%'] + (x_train['training_hours'].describe()['75%'] - x_train['training_hours'].describe()['25%'])*1.5
Min = x_train['training_hours'].describe()['25%'] - (x_train['training_hours'].describe()['75%'] - x_train['training_hours'].describe()['25%'])*1.5
x_train = x_train.loc[(x_train.training_hours >= Min) & (x_train.training_hours <= Max)]
y_train = y_train.loc[y_train.enrollee_id.isin(x_train['enrollee_id'])]

## 불필요 데이터 삭제
x_train.drop(['enrollee_id','city','experience'],axis=1,inplace=True)
y_train.drop(['enrollee_id'],axis=1,inplace=True)
x_test.drop(['enrollee_id','city','experience'],axis=1,inplace=True)
y_test.drop(['enrollee_id'],axis=1,inplace=True)

# 데이터 셋 분리
x_train_train, x_train_test, y_train_train,y_train_test = train_test_split(x_train,y_train,test_size= 0.3, random_state = 23)

# 모델 학습
RFC = RandomForestClassifier(random_state= 23)
RFC.fit(x_train_train,y_train_train)
pred = RFC.predict(x_train_test)
pred_proba = RFC.predict_proba(x_train_test)[:,1]

acc = accuracy_score(y_train_test,pred)
f1 = f1_score(y_train_test,pred)
precision = precision_score(y_train_test,pred)
recall = recall_score(y_train_test,pred)
auc = roc_auc_score(y_train_test,pred_proba)

print(acc)
print(f1)
print(precision)
print(recall)
print(auc)

pred = RFC.predict(x_test)
pred_proba = RFC.predict_proba(x_test)[:,1]

acc = accuracy_score(y_test,pred)
f1 = f1_score(y_test,pred)
precision = precision_score(y_test,pred)
recall = recall_score(y_test,pred)
auc = roc_auc_score(y_test,pred_proba)
x_test['target'] = pred
print(x_test)
print(acc)
print(f1)
print(precision)
print(recall)
print(auc)
```

## 약물 분류 데이터

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler, LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import f1_score, accuracy_score, precision_score, recall_score, roc_auc_score 
from sklearn.ensemble import RandomForestClassifier

#데이터 로드
x_train = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/drug/x_train.csv")
y_train = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/drug/y_train.csv")
x_test= pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/drug/x_test.csv")
y_test= pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/drug/y_test.csv")

print(x_train.describe(include='all'))
print(y_train.describe())

x_train['Sex'] = LabelEncoder().fit_transform(x_train['Sex'])
x_train['BP'] = LabelEncoder().fit_transform(x_train['BP'])
x_train['Cholesterol'] = LabelEncoder().fit_transform(x_train['Cholesterol'])

x_test['Sex'] = LabelEncoder().fit_transform(x_test['Sex'])
x_test['BP'] = LabelEncoder().fit_transform(x_test['BP'])
x_test['Cholesterol'] = LabelEncoder().fit_transform(x_test['Cholesterol'])

x_train.drop(['ID'],axis=1,inplace=True)
x_test.drop(['ID'],axis=1,inplace=True)
y_train.drop(['ID'],axis=1,inplace=True)
y_test.drop(['ID'],axis=1,inplace=True)

x_train_train,x_train_test,y_train_train,y_train_test = train_test_split(x_train,y_train,test_size= 0.33, random_state = 23)

RFC = RandomForestClassifier(random_state = 23)
RFC.fit(x_train_train,y_train_train)
pred = RFC.predict(x_train_test)
pred_proba = RFC.predict_proba(x_train_test)

acc = accuracy_score(y_train_test,pred)
f1 = f1_score(y_train_test,pred,average='macro')
recall = recall_score(y_train_test,pred,average='macro')
precision = precision_score(y_train_test,pred,average='macro')
y_train_test_proba = y_train_test['Drug'].values.ravel()
auc = roc_auc_score(y_train_test_proba,pred_proba, multi_class= 'ovo')

print(acc)
print(f1)
print(recall)
print(precision)
print(auc)

pred = RFC.predict(x_test)
pred_proba = RFC.predict_proba(x_test)

acc = accuracy_score(y_test,pred)

x_test['Drug'] = pred
print(x_test)
f1 = f1_score(y_test,pred,average='macro')
recall = recall_score(y_test,pred,average='macro')
precision = precision_score(y_test,pred,average='macro')
y_test_proba = y_test['Drug'].values.ravel()
auc = roc_auc_score(y_test_proba,pred_proba, multi_class= 'ovo')

print(acc)
print(f1)
print(recall)
print(precision)
print(auc)
```
