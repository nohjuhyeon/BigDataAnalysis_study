# 독립성 검정

```python
import numpy as np
import pandas as pd
from scipy.stats import chi2_contingency
data_matrix = np.array([[50,75,125,175],[90,30,45,10]])
chi2,p,df,expected = chi2_contingency(data_matrix)

print('chi-square statistic : ',chi2)
print('p-value : ',p)
print('Degrees of freedom : ',df)
print('Expected frequencies table: ')
print(expected)

if p > 0.05:
    print('귀무가설을 기각할 수 없습니다. 두 변수는 독립적이며, 연관성이 없습니다.')
else:
    print('귀무가설을 기각합니다. 두 변수는 독립적이지 않으며, 연관성이 있습니다.')
```

```python
import numpy as np
import pandas as pd
from scipy.stats import chi2_contingency
df = pd.read_csv("https://raw.githubusercontent.com/YoungjinBD/dataset/main/titanic.csv")
cross_table = pd.crosstab(df['Sex'],df['Survived'])
chi2,p,df,expected = chi2_contingency(cross_table)
print('chi-square statistic : ',chi2)
print('p-value : ',p)
print('Degrees of freedom : ',df)
print('Expected frequencies table: ')
print(expected)

if p > 0.05:
    print('귀무가설을 기각할 수 없습니다. 두 변수는 독립적이며, 연관성이 없습니다.')
else:
    print('귀무가설을 기각합니다. 두 변수는 독립적이지 않으며, 연관성이 있습니다.')
```

```python
import pandas as pd
import numpy as np
from scipy.stats import chi2_contingency
df_titanic = pd.read_csv("https://raw.githubusercontent.com/YoungjinBD/dataset/main/titanic.csv")
print(df_titanic['Age'].isnull().sum())
df_titanic['Age'].fillna(df_titanic['Age'].mean(),inplace=True)
print(df_titanic['Age'].isnull().sum())
print(df_titanic['Age'].describe())
def grouping(row):
    if row < 20:
        answer = '10대'
    elif row <30:
        answer = '20대'
    elif row <40:
        answer = '30대'
    elif row <50:
        answer = '40대'
    elif row <60:
        answer = '50대'
    else:
        answer = '60대 이상'
    return answer
df_titanic['Age_group'] = df_titanic['Age'].apply(grouping)
cross_table = pd.crosstab(df_titanic['Age_group'],df_titanic['Survived'])
chi2,p,df,expect = chi2_contingency(cross_table)
print('chi-squared statistic : ',chi2)
print('p-value : ',p)
print('degrees of freedom : ',df)
print('기대값(Expected of frequenies table) : ')
print(expected)
if p > 0.05:
    print('귀무가설을 기각할 수 없습니다. 두 변수는 독립적이며, 연관성이 없습니다.')
else:
    print('귀무가설을 기각합니다. 두 변수는 연관성이 있습니다.')
```
